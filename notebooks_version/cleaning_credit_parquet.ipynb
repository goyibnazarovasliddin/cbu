```python
import pandas as pd

credit = pd.read_parquet("credit_history.parquet")
credit.head()
```

```python
credit.info()
```

```python
credit['num_delinquencies_2yrs'].value_counts()
```

```python
credit['num_delinquencies_2yrs'] = credit['num_delinquencies_2yrs'].fillna(0)
```

```python
credit.isnull().sum()
```

```python
credit['num_public_records'].value_counts()
```

```python
credit = credit.drop('num_public_records',axis=1)
credit.head()
```

```python
credit['num_collections'].value_counts()
```

```python
credit = credit.drop('num_collections',axis=1)
credit.head()
```

```python
credit.rename(columns={'customer_number':'customer_id'}, inplace=True)
```

```python
credit.to_csv("cleaned_credit_history.csv", index=False)
```
