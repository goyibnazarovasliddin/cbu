```python
import numpy as np 
import pandas as pd 
```

```python
df_ratios = pd.read_json("financial_ratios.jsonl" , lines = 'True')

# Birinchi 5 qatorni ko'rish
df_ratios.head()
```

```python
df_ratios.info()
```

```python
object_cols = ['monthly_income', 'existing_monthly_debt', 'monthly_payment',
               'revolving_balance', 'credit_usage_amount', 'available_credit',
               'total_monthly_debt_payment', 'total_debt_amount', 'monthly_free_cash_flow']

for col in object_cols:
    # $ va , belgilarini olib tashlash va float ga oâ€˜tkazish
    df_ratios[col] = df_ratios[col].replace('[\$,]', '', regex=True).astype(float)

df_ratios.head()
```

```python
df_ratios.info()
```

```python
df_ratios.rename(columns={'cust_num':'customer_id'}, inplace=True)
```
