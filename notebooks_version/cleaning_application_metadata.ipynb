```python
df = pd.read_csv('application_metadata.csv')
df.head() 
```

```python
df.rename(columns={'customer_ref':'customer_id'}, inplace=True)
```

```python
df2 = df.drop('random_noise_1', axis=1)
df2.head()
```

```python
df2 = df2.drop('application_hour',axis=1)
df2.head()
```

```python
df2 = df2.drop('application_day_of_week',axis=1)
df2.head()
```

```python
df2 = df2.drop('preferred_contact',axis=1)
df2.head()
```

```python
df2.info()
```

```python
from sklearn.preprocessing import LabelEncoder
categorical_cols = ['referral_code', 'account_status_code']
for col in categorical_cols:
    le = LabelEncoder()
    df2[col] = le.fit_transform(df2[col].astype(str))
```
