```python
import pandas as pd 
df2 = pd.read_csv('demographics.csv')
df2.head()
```

```python
df2.info()
```

```python
df2.rename(columns={'cust_id':'customer_id'}, inplace=True)
```

```python
df2.head()
```

```python
df2['age'].value_counts()
```

```python
df2['annual_income'] = df2['annual_income'].replace('[\$,]', '', regex=True).astype(float)
df2.info()
```

```python
df2['employment_type'].unique()
```

```python
# Employment type mapping
employment_map = {
    'Full-time': 'Full-time', 'FULL_TIME': 'Full-time', 'Full Time': 'Full-time', 'Fulltime': 'Full-time', 'FT': 'Full-time',
    'Part Time': 'Part-time', 'PART_TIME': 'Part-time', 'PT': 'Part-time', 'Part-time': 'Part-time',
    'Self Employed': 'Self-employed', 'Self-employed': 'Self-employed', 'SELF_EMPLOYED': 'Self-employed',
    'Contractor': 'Contract', 'Contract': 'Contract', 'CONTRACT': 'Contract'
}

# Apply mapping
df2['employment_type'] = df2['employment_type'].map(employment_map)
df2['employment_type'].unique()
```

```python
mode_type = df2['employment_type'].mode()[0]
df2['employment_type'].fillna(mode_type, inplace=True)
df2['employment_type'].unique()
```

```python
df2['education'].unique()
```

```python
df2['marital_status'].unique()
```

```python
df2['num_dependents'].value_counts()
```

```python
mode_type = df2['employment_length'].mode()[0]
df2['employment_length'].fillna(mode_type, inplace=True)
df2.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
categorical_cols = ['employment_type', 'education', 'marital_status']
for col in categorical_cols:
    le = LabelEncoder()
    df2[col] = le.fit_transform(df2[col].astype(str))
df2.head()
```

```python
df2.to_csv("cleaned_demographic.csv", index=False)
```
